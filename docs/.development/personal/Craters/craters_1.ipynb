```python
from redplanet import utils

import pandas as pd

path__database = r'C:\Users\Eris\Downloads\redplanet-data\Craters\Catalog_Mars_Release_2020_1kmPlus_FullMorphData.csv'

config = [
    ['CRATER_ID', 'ID', str],
    ['LAT_CIRC_IMG', 'lat', float],
    ['LON_CIRC_IMG', 'clon', float],
    ['DIAM_CIRC_IMG', 'diameter_km', float],
]

database1 = pd.read_csv(
    path__database, 
    usecols=[x[0] for x in config],
    dtype={x[0]: x[2] for x in config}
)

database1 = database1.rename(columns={x[0]: x[1] for x in config})

database1['lon'] = utils.clon2lon(database1['clon'])

database1 = database1[database1['diameter_km'] > 100]
database1 = database1.to_dict('index')
# utils.print_dict(database1)
database1
```
